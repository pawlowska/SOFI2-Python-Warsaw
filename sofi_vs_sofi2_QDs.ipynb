```python
import os, tifffile
import numpy as np
import matplotlib.pyplot as plt
import sofi
import SOFI2_0_fromMatlab as sofi2

plt.rcParams['figure.figsize'] = [8, 8]
```

```python
# read data and show mean

fname='QD_series_1'
data_dir='pliki/2020_02_21-QDs/'
data_path =data_dir+fname

data=np.array(tifffile.imread(os.path.join(data_dir, fname+'.tif')))
data_mean=data.mean(axis=0)
plt.imshow(data_mean)
```

```python
# 2nd autocumulant
ac2=sofi.autocumulant_2(data, 1)
plt.imshow(ac2)
```

```python
#try ldrc on ac2
ac2_ldrc=sofi2.ldrc(ac2, data_mean, 15)
plt.imshow(ac2_ldrc)
```

```python
# 4th autocumulant
ac4=sofi.autocumulant_4(data, 1, 2, 3)
plt.imshow(abs(ac4))
```

```python
#m4 & ldrc
m4_ldrc=sofi2.ldrc(sofi2.M4(data), data_mean, 15)
plt.imshow(m4_ldrc)
```

```python
#m6 & ldrc
m6=sofi2.M6(data)
m6_ldrc=sofi2.ldrc(m6, data_mean, 15)
plt.imshow(m6_ldrc)
```
