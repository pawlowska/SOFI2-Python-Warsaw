```python
import os, tifffile
import numpy as np
import matplotlib.pyplot as plt
import sofi
import SOFI2_0_fromMatlab as sofi2
```

```python
# read data and show mean

fname='SRRF-high_1'
data_dir='SRRF-github/'
data_path =data_dir+fname

data=np.array(tifffile.imread(os.path.join(data_dir, fname+'.tif')))
data_mean=data.mean(axis=0)
plt.imshow(data_mean)
plt.colorbar()
```

```python
# 2nd autocumulant
ac2=sofi.autocumulant_2(data, 1)
plt.imshow(ac2)
plt.colorbar()
```

```python
#try ldrc on ac2
ac2_ldrc=sofi2.ldrc(ac2, data_mean, 15)
plt.imshow(ac2_ldrc)
plt.colorbar()
```

```python
# 4th autocumulant
ac4=sofi.autocumulant_4(data, 1, 2, 3)
plt.imshow(ac4)
plt.colorbar()
```

```python
#m2 & ldrc
m2_ldrc=sofi2.ldrc(sofi.autocumulant_2(data,0), data_mean, 15)
plt.imshow(m2_ldrc)
plt.colorbar()
```

```python
#m4 & ldrc
m4_ldrc=sofi2.ldrc(sofi.autocumulant_4(data,0,0,0), data_mean, 15)
plt.imshow(m4_ldrc)
plt.colorbar()
```

```python
#m6 & ldrc
m6=sofi2.M6(data)
m6_ldrc=sofi2.ldrc(m6, data_mean, 15)
plt.imshow(m6_ldrc)
plt.colorbar()
```
